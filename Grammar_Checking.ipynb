```python
# Install necessary libraries
!pip install transformers gradio

# Import libraries
from transformers import pipeline
import gradio as gr

# Load a pre-trained masked language model
model = pipeline('fill-mask', model='bert-base-uncased')

# Function to perform basic grammar correction using masking
def grammar_check(text):
    words = text.split()
    corrected_text = text

    # Mask each word one by one and get the model's prediction
    for i in range(len(words)):
        masked_sentence = ' '.join(words[:i] + ['[MASK]'] + words[i+1:])
        result = model(masked_sentence)
        predicted_word = result[0]['token_str']
        words[i] = predicted_word

    corrected_text = ' '.join(words)
    return corrected_text

# Create Gradio interface
interface = gr.Interface(
    fn=grammar_check,
    inputs="text",
    outputs="text",
    title="Grammar Checker",
    description="Enter text to check for grammar mistakes."
)

# Launch the interface
interface.launch()
```
